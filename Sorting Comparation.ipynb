```python
#Como ejercicio se deja demostrar que el orden de crecimiento de la función cocktailsort es de O(n^2)
#Como ejercicio se deja demostrar que el orden de crecimiento del insertion-sort Peor Caso:O(n^2) - Mejor Caso: O(n)

from random import * ##Librería para generar números aleatorios en python
from time import *   #Librería para manejo de tiempos de ejecución en python
import matplotlib.pyplot as plt #Librería para generar gráficos en python


def generateData(n):#funcion para generar un arreglo con valores aleatorios de n posiciones: O(n)
    dataArray=[]
    for i in range(0, n):
        dataArray.append(randint(0, n)) 
    return dataArray

#Función de ordenamiento bubblesort o burbuja
#                                             Peor Caso                 Mejor Caso
#                                         El arreglo este ordenado  El arreglo este
#                                         en sentido inverso        ordenado
def bubblesort(arre):                     # veces   costo       # veces   costo
    temp=0                                #   1       c1            1       c1
    for i in range(len(arre)-1):          #  n+1      c2           n+1      c2
        for j in range(i+1,len(arre)):    #∑(n+1-i)   c3        ∑(n+1-i)    c3  Para i=1, hasta n+1
            if (arre[i]>arre[j]):         #∑(n+1-i)   c4        ∑(n+1-i)    c4  Para i=1, hasta n
                temp=arre[i]              #∑(n+1-i)   c5            0       c5  Para i=1, hasta n 
                arre[i]=arre[j]           #∑(n+1-i)   c6            0       c6  Para i=1, hasta n    
                arre[j]=temp              #∑(n+1-i)   c7            0       c7  Para i=1, hasta n 

#Queda como ejercicio demostrar 
#∑(n+1-i) = n^2 + n  + n +1 - (n^2)/2 - (3n/2) -2 , para i=1 hasta n+1
#           ____________________________________
#                    (n^2)/2 + (n/2) - 1
#
#∑(n+1-i) = n^2 + n - (n^2/2) - (n/2), para i=1 hasta n
#           ____________________________________
#                    (n^2)/2 + (n/2) 
#Y demuestre que el bubble sort en el peor y mejor caso tiene un orden de crecimiento de  O(n^2)


#Función de ordenamiento cocktail sort               
def cocktailSort(arre):
    up = range(len(arre)-1)
    while True:
        for indices in (up, reversed(up)):
            swapped = False
            for i in indices:
                if arre[i] > arre[i+1]:  
                    arre[i], arre[i+1] =  arre[i+1], arre[i]
                    swapped = True
            if not swapped:
                return

#Función de ordenamiento insertion sort            
def insertion_sort(arre):
    for i in xrange(1, len(arre)):
        j = i-1 
        key = arre[i]
        while (arre[j] > key) and (j >= 0):
           arre[j+1] = arre[j]
           j -= 1
        arre[j+1] = key
    
    

seed()#Si inicializa la semilla para la generación de números aleatorios
timesBubble=[]#Se inicializa el arreglo que almacenará los tiempos de bubble sort
timesCockTail=[]#Se inicializa el arreglo que almacenará los tiempos de cocktail sort
timesInsertion=[]#Se inicializa el arreglo que almacenará los tiempos de insertion sort
ndata=[]#Se inicializa el arreglo que almacenará los datos(100,200,300,400,....,4900,5000) del eje x


for k in range(100,5001,100):
    ar = generateData(k)#Aquí se generan números en orden aleatorio entre 0 y k (donde k es el tamaño de muestro en cada iteración)
                        #para el algoritmo de bubble sort
    start = time()#Se inicia el cronómetro de tiempo
    bubblesort(ar)#Se invoca la función de bubblesort crecimiento de O(n^2)
    timesBubble.append(round(time() - start,2))#Se almacenan el tiempo transcurrido desde el inicio hasta el final del
                                               #ordenamiento por bubble sort, con dos decimales
    
    
    ar = generateData(k)#Aquí se generan números en orden aleatorio entre 0 y k (donde k es el tamaño de muestro en cada iteración)
                        #para el algoritmo de cocktail sort
    
    start = time()#Se inicia el cronómetro de tiempo
    cocktailSort(ar)#Se invoca la función de cocktailsort crecimiento de O(n^2)
    timesCockTail.append(round(time() - start,2))#Se almacenan el tiempo transcurrido desde el inicio hasta el final del
                                                #ordenamiento por cocktail sort con dos decimales
    
    
    ar = generateData(k)#Aquí se generan números en orden aleatorio entre 0 y k (donde k es el tamaño de muestro en cada iteración)
                        #para el algoritmo de insertion sort
    start = time()#Se inicia el cronómetro de tiempo
    insertion_sort(ar)#Se invoca la función de insertionsort crecimiento de O(n^2)
    timesInsertion.append(round(time() - start,2))#Se almacenan el tiempo transcurrido desde el inicio hasta el final del
                                                #ordenamiento por insertion sort con dos decimales
    
    
    ndata.append(k)#Aquí se va guardando los valores del eje x (tamaño de muestra)
    

fig = plt.figure(figsize=(14, 7), dpi=100)#Dimension de la figura    
plt.plot(ndata,timesBubble,'r')#Aquí se realiza una gráfica de los tiempos de bubble sort, color rojo(r=red)
plt.plot(ndata,timesCockTail,'g')#Aquí se realiza una gráfica de los tiempos de cocktail sort, color verde(g=green)
plt.plot(ndata,timesInsertion,'b')#Aquí se realiza una gráfica de los tiempos de bubble sort, color azul(b=blue)

plt.title("Algoritmos de ordenamiento")
plt.xlabel("Datos (n)")#Eje X
plt.ylabel("Tiempo (s)")#Eje Y
plt.legend(["Bubble Sort","CockTail Sort","Insertion Sort"],loc=2)#Leyendas de las gráficas

#Configuración eje X
#plt.xlim(0,500)

#Configuración eje Y
#plt.ylim(0,10)

plt.grid()#Dibuja la grilla
plt.show()#Dibujar las gráficas

```
