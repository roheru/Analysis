```python

from random import * ##Librería para generar números aleatorios en python
from time import *   #Librería para manejo de tiempos de ejecución en python
import matplotlib.pyplot as plt #Librería para generar gráficos en python


def generateData(n):#funcion para generar un arreglo con valores aleatorios de n posiciones: O(n)
    dataArray=[]
    for i in range(0, n):
        dataArray.append(randint(0, n)) 
    return dataArray



def merge_sort(m):
    if len(m) <= 1:
        return m
 
    middle = len(m) / 2
    left = m[:middle]
    right = m[middle:]
 
    left = merge_sort(left)
    right = merge_sort(right)
    return list(merge(left, right))



def merge(left, right):
    result = []
    left_idx, right_idx = 0, 0
    while left_idx < len(left) and right_idx < len(right):
        # change the direction of this comparison to change the direction of the sort
        if left[left_idx] <= right[right_idx]:
            result.append(left[left_idx])
            left_idx += 1
        else:
            result.append(right[right_idx])
            right_idx += 1
 
    if left:
        result.extend(left[left_idx:])
    if right:
        result.extend(right[right_idx:])
    return result

def partition(list, start, end):
    pivot = list[end]                          
    bottom = start-1                           
    top = end                                  

    done = 0
    while not done:                            

        while not done:                        
            bottom = bottom+1                  

            if bottom == top:                  
                done = 1                       
                break

            if list[bottom] > pivot:           
                list[top] = list[bottom]       
                break                          

        while not done:                        
            top = top-1                        
            
            if top == bottom:                  
                done = 1                       
                break

            if list[top] < pivot:              
                list[bottom] = list[top]       
                break                          

    list[top] = pivot                          
    return top                                 


def quicksort(list, start, end):
    if start < end:                            
        split = partition(list, start, end)    
        quicksort(list, start, split-1)        
        quicksort(list, split+1, end)
    else:
        return

seed()#Si inicializa la semilla para la generación de números aleatorios
timesMerge=[]#Se inicializa el arreglo que almacenará los tiempos de Merge sort
timesQuick=[]#Se inicializa el arreglo que almacenará los tiempos de Quick sort
ndata=[]#Se inicializa el arreglo que almacenará los datos(100,200,300,400,....,4900,5000) del eje x


for k in range(1000,300000,5000):
    ar = generateData(k)#Aquí se generan números en orden aleatorio entre 0 y k (donde k es el tamaño de muestro en cada iteración)
                        #para el algoritmo de bubble sort
    start = time()#Se inicia el cronómetro de tiempo
    merge_sort(ar)#Se invoca la función de bubblesort crecimiento de O(n^2)
    timesMerge.append(round(time() - start,2))#Se almacenan el tiempo transcurrido desde el inicio hasta el final del
                                               #ordenamiento por bubble sort, con dos decimales
    
    
    ar = generateData(k)#Aquí se generan números en orden aleatorio entre 0 y k (donde k es el tamaño de muestro en cada iteración)
                        #para el algoritmo de cocktail sort
    
    start = time()#Se inicia el cronómetro de tiempo
    quicksort(ar,0,len(ar)-1)#Se invoca la función de cocktailsort crecimiento de O(n^2)
    timesQuick.append(round(time() - start,2))#Se almacenan el tiempo transcurrido desde el inicio hasta el final del
                                                #ordenamiento por cocktail sort con dos decimales
    
    ndata.append(k)#Aquí se va guardando los valores del eje x (tamaño de muestra)
    

fig = plt.figure(figsize=(14, 7), dpi=100)#Dimension de la figura    
plt.plot(ndata,timesMerge,'r')#Aquí se realiza una gráfica de los tiempos de bubble sort, color rojo(r=red)
plt.plot(ndata,timesQuick,'g')#Aquí se realiza una gráfica de los tiempos de cocktail sort, color verde(g=green)


plt.title("Algoritmos de ordenamiento")
plt.xlabel("Datos (n)")#Eje X
plt.ylabel("Tiempo (s)")#Eje Y
plt.legend(["Merge Sort","Quick Sort"],loc=2)#Leyendas de las gráficas

#Configuración eje X
#plt.xlim(0,500)

#Configuración eje Y
#plt.ylim(0,10)

plt.grid()#Dibuja la grilla
plt.show()#Dibujar las gráficas

```
